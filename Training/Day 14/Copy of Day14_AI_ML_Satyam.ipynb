```python
import random
intents = {
    "greeting": ["Hi there!", "How can I help you?", "Hello!", "Hey!"],
    "goodbye": ["Goodbye!", "See you later!", "Bye"],
    "thankyou": ["You're welcome!", "No problem!", "Thank you!"],
    "address": ["IITM college near Tihar Jail"],
    "courses": ["IITM college offers 5 graduation courses"]
}
```

```python
def detect_response(query):
  prediction = humingbird.Text.predict(
    text = query,
    labels = ["greeting","goodbye",'thankyou',"address","courses"]
  )
  max_score = 0
  for val in prediction:
    if val["score"] > max_score:
      max_score = val["score"]
      label = val["className"]
  print(label," : ",max_score)
  return random.choice(intents[label])
```

```python
import random
intents = {
    "greeting": ["Hi there!", "Hello!", "Hey!"],
    "goodbye": ["Goodbye!", "See you later!", "Bye"],
    "thankyou": ["You're welcome!", "No problem!", "Thank you!"],
    "bank address": ["Bank of Delhi, Janakpuri, Near Tihar Jail, New Delhi"],
    "Balance": ["As an AI model I don't have your personal information, you can check your balance by logging into your account"],
    "Loans" : ["Home Loan, Personal Loan, Education Loan"],
    "Home Loan": ["A home loan is like borrowing money from a bank to buy a house. Instead of paying the entire price upfront, you pay a little bit each month (called EMI). Over time, you pay back the loan along with interest."],
    "Personal Loan": ["A personal loan is like borrowing money from a bank or financial institution for your personal needs. You get a lump sum amount and then pay it back in fixed installments over a set period."],
    "Education Loan": ["An education loan is like borrowing money from a bank or financial institution for your education needs. You get a lump sum amount and then pay it back in fixed installments over a set period."],
    "Home Loan Interest": ["8.70% p.a. onwards"],
    "Personal Loan Interest": ["6.40% p.a. onwards"],
    "Education Loan Interest": ["4.00% p.a. onwards"]
}
```

```python
def detect_response(query):
  prediction = humingbird.Text.predict(
    text = query,
    labels = ["greeting","goodbye",'thankyou',"bank address","Balance","Loans","Home Loan","Personal Loan","Education Loan","Home Loan Interest","Personal Loan Interest","Eduction Loan Interest"]
  )
  max_score = 0
  for val in prediction:
    if val["score"] > max_score:
      max_score = val["score"]
      label = val["className"]
  print(label," : ",max_score)
  if label == "greeting":
    return random.choice(intents[label]+[" How can I help you?"])
  return random.choice(intents[label])
```
