```python
import numpy as np
np.random.seed(3)
A = np.random.randint(0,100,size=(9,9))
print("Initial Matrix")
print(A)
B = []
C = np.zeros((A.shape[0], A.shape[1]))
for row in range(A.shape[0]):
  mean = (A[row,:]).mean()
  B.append(mean)

zerosCount = 0
nonZerosCount = 0

for i in range(A.shape[0]):
    for j in range(A.shape[1]):
        if B[i] < A[i,j]:
            C[i,j] = 0
            zerosCount = zerosCount+1
        else:
            C[i,j] = A[i,j]
            nonZerosCount = nonZerosCount+1

print("Number of zeros: ",zerosCount)
print("Number of non-zeros: ",nonZerosCount)

if(nonZerosCount+zerosCount == A.size):
    print("Updated Matrix is verified")

print("Updated Matrix")
print(C)
```
