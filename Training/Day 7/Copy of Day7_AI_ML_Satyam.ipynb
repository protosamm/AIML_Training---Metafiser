```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn import model_selection
from sklearn import metrics
from sklearn import tree

data = datasets.load_iris()

x = data.data
y = data.target

ratio = 0.3
xtrain,xtest,ytrain,ytest = model_selection.train_test_split(x,y,test_size=ratio, random_state=4)

#train the model

tree_model = tree.DecisionTreeClassifier(criterion='entropy',max_depth=None)
tree_model = tree_model.fit(xtrain,ytrain)

#predict training data for training accuracy
yPredTrain = tree_model.predict(xtrain)

#predict testing data for testing accuracy
yPredTest = tree_model.predict(xtest)

accTrain = metrics.accuracy_score(yPredTrain,ytrain)

accTest = metrics.accuracy_score(yPredTest,ytest)

print("Training accuracy: ",accTrain)
print("Testing accuracy: ",accTest)

plt.figure(1)
tree.plot_tree(tree_model,filled=True)
```

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn import model_selection
from sklearn import metrics
from sklearn import tree

data = pd.read_csv('/content/diabetes.csv')

x = data.drop("Outcome", axis=1, inplace=False)
y = data['Outcome']

ratio = 0.3
xtrain,xtest,ytrain,ytest = model_selection.train_test_split(x,y,test_size=ratio, random_state=4)

#train the model

tree_model = tree.DecisionTreeClassifier(criterion='gini',max_depth=6)
tree_model = tree_model.fit(xtrain,ytrain)

#predict training data for training accuracy
yPredTrain = tree_model.predict(xtrain)

#predict testing data for testing accuracy
yPredTest = tree_model.predict(xtest)

accTrain = metrics.accuracy_score(yPredTrain,ytrain)

accTest = metrics.accuracy_score(yPredTest,ytest)

print("Training accuracy: ",accTrain)
print("Testing accuracy: ",accTest)

plt.figure(1)
tree.plot_tree(tree_model, filled=True)
plt.show()
```

```python
import matplotlib.image as mimg
import numpy as np
import matplotlib.pyplot as plt
total_samples = 410
data = np.zeros((total_samples,112*92))
label = np.zeros((total_samples))
images = np.zeros((total_samples,112,92))

index = -1
for i in range(1,42,1):
  for j in range(1,11,1):
    user_id = i
    samp_no = j
    index = index + 1
    path = "/gdrive/MyDrive/AI_Summer_Training/orl_face/u%d/%d.png"%(user_id, samp_no)

    im = mimg.imread(path)
    feat = im.reshape(1,-1)
    data[index,:] = feat
    label[index] = i
    images[index,:,:] = im

print(images.shape)

```

```python
import numpy as np
# visualise the data images
plt.figure(1,(10,10))
for i in range(5):
  for j in range(5):
    num = np.random.randint(0,410)
    samples_im = images[num,:,:]
    plt.subplot(5,5,(i+j*5)+1)
    plt.imshow(samples_im,cmap='gray')
    plt.title(str(int(label[num])))
    plt.axis('off')
```

```python
from sklearn import svm,model_selection,metrics
import numpy as np
import pandas as pd

x = data
y = label

xtrain,xtest,ytrain,ytest = model_selection.train_test_split(x,y,test_size=0.3,random_state=4)

kernel = ["poly", "rbf", "linear"]
Cvalue = [1,2,3]

result = np.zeros((len(kernel),len(Cvalue)))
for i in range(len(kernel)):
  for j in range(len(Cvalue)):
    img_model = svm.SVC(kernel=kernel[i], gamma="scale",C=Cvalue[i])
    img_model = img_model.fit(xtrain,ytrain)
    ypred = img_model.predict(xtest)

    acc = metrics.accuracy_score(ypred,ytest)
    result[i,j]=acc

ResultDF = pd.DataFrame(result,index=kernel,columns=["C=1","C=2","C=3"])
ResultDF = ResultDF.T
print(ResultDF)
ResultDF.plot(kind='bar',figsize=(4,4))
```

```python
#Saving our trained model in google drive
#we use joblib to save models

import joblib
from sklearn import svm
img_model = svm.SVC(kernel='linear', gamma="scale",C=1)
img_model = img_model.fit(xtrain,ytrain)
joblib.dump(img_model,'/gdrive/MyDrive/AI_Summer_Training/data_files/orl_face_model2.pkl')
```
